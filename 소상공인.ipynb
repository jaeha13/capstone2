## 관악구

```python
lat = data[data['시군구명'] == '관악구']['위도'].to_list()
long = data[data['시군구명'] == '관악구']['경도'].to_list()
```

```python
import folium
from folium.plugins import HeatMap
```

```python
jsonlat=[]
jsonlng=[]
for lt, lng in zip(lat, long):
    jsonlat.append(lt)
    jsonlng.append(lng)
```

```python
def allroute():
    m = folium.Map(location=(37.46681023316536, 126.94493372385809), zoom_start=14)
    
    heatMap = HeatMap(zip(jsonlat, jsonlng),
                   min_opacity=0.1,
                   max_val=5,
                   radius=10, blur=15,
                   max_zoom=5, color='red')

    m.add_child(heatMap)
    
    return m
```

```python
allroute()
```

```python
allroute().save('./result/관악구 상권 분포 히트맵.html')
```

### restaurant

```python
rest = pd.read_csv('./result/restaurant.csv')
```

```python
grouped_r = rest.groupby('시군구명')[['시군구명']].count()
grouped_r.columns = ['한식점 개수']
grouped_r.reset_index(inplace=True)
```

```python
grouped_r.columns = ['자치구', '한식점 개수']
```

```python
grouped_r.to_csv('./result/구별 한식점.csv', index=False)
```

```python
import folium
import json
#geo_path = 'data/seoul_geo.json'
geo_path = '../data/seoul.geojson'
geo_str = json.load(open(geo_path, encoding='utf-8'))
#print(geo_str)
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11)
fmap=folium.Choropleth(geo_data = geo_str,
               data = grouped_r,
               columns = ['자치구', '한식점 개수'],
               fill_color = 'YlOrRd',
               key_on='feature.properties.name').add_to(map)
fmap.geojson.zoom_on_click = False
fmap.geojson.add_child(
    folium.features.GeoJsonTooltip(['name'],labels=False)
)

map.save('./result/서울 한식점 점포.html')
display(map)
```
